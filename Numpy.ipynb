### 배열 값 삭제
- delete() 배열의 특정 위치에 값 삭제

### 배열 복사

### 배열 분할

### 불리언 연산자

```python
a2= np.arange(1,10).reshape(3,3)
print(a2)
print((a2>5)&(a2<8))
print(a2[(a2>5)&(a2<8)])

print((a2>5)|(a2<8))
print(a2[(a2>5)|(a2<8)])

print((a2>5)^(a2<8))
print(a2[(a2>5)^(a2<8)])

print(~(a2>5) )
print(a2[~(a2>5)])
```
